```r
# Set the seed for reproducibility
set.seed(123)

```

```r
# Generate random data
x <- rnorm(100)  # 100 random numbers for x
y <- rnorm(100)  # 100 random numbers for y

```

```r
# Create a scatter plot
plot(x, y, 
     main = "Scatter Plot of Random Data", 
     xlab = "X Values", 
     ylab = "Y Values", 
     col = "blue", 
     pch = 19)  # 'pch' changes the shape of the points
```
